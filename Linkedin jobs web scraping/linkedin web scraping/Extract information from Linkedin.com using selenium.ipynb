### Extract information from Linkedin.com using selenium

```python
#Importing libraries
import pandas as pd
import selenium
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
#Assigning the webdrive to the url
browser=webdriver.Chrome(r"C:\Users\DA\Desktop\selenium\chromedriver.exe")
browser.get('https://www.linkedin.com')
```

```python
#Entering my account on linkedin
time.sleep(3)
username=browser.find_element_by_id("session_key")
username.send_keys('abdallahkhaledellawah@gmail.com')
password=browser.find_element_by_id('session_password')
password.send_keys('620620620')# i had to hide the password since i am uploading the project on my github profile
password.send_keys(Keys.RETURN)
```

```python
browser.get('https://www.linkedin.com/jobs/')
time.sleep(3)
# find the keywords/location search bars:
search_bars = browser.find_elements_by_class_name('jobs-search-box__text-input')
search_keywords = search_bars[0]
search_keywords.send_keys('Data analyst')

search_location = search_bars[3]
search_location.send_keys('Dubai, United Arab Emirates')
search_location.send_keys(Keys.RETURN)
```

```python
# get a list of all the listings elements's in the side bar
list_items = browser.find_elements_by_class_name("occludable-update")
lists=[]
# scrolls a single page:
for job in list_items:
    browser.execute_script("arguments[0].scrollIntoView();", job)
    job.click()
    lists.append(job.text.split('\n')[:3])
```

```python
lists
```

```python
#puttin info into seperate lists
Job_title=([ele[0] for ele in lists])
Company=([ele[1] for ele in lists])
location=([ele[2:] for ele in lists])
```

```python
#put our info in a dictionary
dict = {'Job_title': Job_title, 'Company': Company, 'location': location } 
```

```python
#Convert the dictionary into a dataframe
df = pd.DataFrame(dict)
```
